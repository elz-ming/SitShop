## #2.2 Visualize the normalized rating

```python
mean_rating_df = user_table[['username','mean_rating']]
newReview_table = review_table.merge(mean_rating_df, on='username', how='left')
newReview_table['normalized_rating'] = newReview_table['rating'] - newReview_table['mean_rating']
display(newReview_table.head(3))
```

```python
aggregate = {
    "normalized_rating": 'mean',
    "rating"           : 'mean'
}

rename_dict = {
    'normalized_rating'    :"product_mean_normalized_rating",
    'rating'               :"product_mean_rating"
}

normalize_rating_df = newReview_table.groupby('product_id').agg(aggregate).reset_index()
normalize_rating_df = normalize_rating_df.rename(columns=rename_dict)
normalize_rating_df['product_mean_normalized_rating'] = normalize_rating_df['product_mean_normalized_rating'].round(2)
normalize_rating_df['product_mean_rating'] = normalize_rating_df['product_mean_rating'].round(2)

display(normalize_rating_df.head(3))
sns.scatterplot(data=normalize_rating_df, x='product_mean_rating', y='product_mean_normalized_rating')
plt.show()
```

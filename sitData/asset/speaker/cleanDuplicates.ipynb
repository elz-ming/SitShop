```python
import pandas as pd
```

```python
# Read the csv file and save it as df
df1 = pd.read_csv('keyboardProduct.csv')
df2 = pd.read_csv('keyboardMerchant.csv')
df3 = pd.read_csv('keyboardReview.csv')
```

```python
df1
```

```python
# Remove duplicate rows based on both 'username' and 'product_id'
product_df = df1.drop_duplicates(subset=['description', 'total_sold']).reset_index(drop=True)

#product_df.to_csv("keyboardProduct.csv")
product_df
```

```python
merchant_iddf = list(product_df['merchant_id'])
mask = df2['merchant_id'].isin(merchant_iddf)

# Filter the DataFrame based on the mask
filtered_data2 = df2[mask].reset_index(drop=True)
filtered_data2
```

```python
product_iddf = list(product_df['product_id'])
mask = df3['product_id'].isin(product_iddf)

# Filter the DataFrame based on the mask
filtered_data3 = df3[mask].reset_index(drop=True)
filtered_data3
```

```python
product_df.to_csv("product.csv", index=False)
filtered_data2.to_csv("merchant.csv", index=False)
filtered_data3.to_csv("review.csv", index=False)
```
